```python
import pandas as pd
```

```python
ruta1 = ('C:/DEV/Proyectos/Analisis_Ventas/Datos_Ventas_Tienda.csv')
ruta2 = ('C:/DEV/Proyectos/Analisis_Ventas/Datos_Ventas_Tienda2.csv')
```

```python
df1 = pd.read_csv(ruta1)
df2 = pd.read_csv(ruta2)
```

# 1. Juntamos los dos CSV en un DataFrame completo

```python
df_completo = pd.concat([df1, df2], ignore_index= True)
df_completo
```

```python
type(df_completo['Fecha'][0])
```

```python
# Comprobamos que en este archivo no hay nulos
df_completo.isnull().sum()
```

# 2. Transformamos el tipo de datos de fecha

```python
df_completo['Fecha'] = pd.to_datetime(df_completo['Fecha'])
df_completo
```

```python
type(df_completo['Fecha'][0])
```

# 3. Averiguamos el producto más vendido

```python
producto_mas_vendido = df_completo.groupby('Producto')['Cantidad'].sum()
producto_mas_vendido = producto_mas_vendido.sort_values(ascending=False).head(1)
producto_mas_vendido
```

```python
por_mes = df_completo['Fecha'].dt.month
por_mes
```

# 4. Sacamos el mes donde ha habido más ventas

```python
mes_con_mas_ventas = df_completo.groupby(por_mes)['Total Venta'].sum()
mes_con_mas_ventas = mes_con_mas_ventas.sort_values(ascending=False).head(1)
mes_con_mas_ventas
```

# 5. Agrupamos por tipo de producto cual ha tenido mayor éxito

```python
ventas_por_categoria = df_completo.groupby('Producto')['Total Venta'].sum()
ventas_por_categoria
```
